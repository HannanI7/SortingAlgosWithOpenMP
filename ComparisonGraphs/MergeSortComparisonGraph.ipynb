```python
import matplotlib.pyplot as plt
import numpy as np

!rm -rf SortingAlgosWithOpenMP

!git clone https://github.com/HannanI7/SortingAlgosWithOpenMP

# Navigate to the directory containing the C files
%cd SortingAlgosWithOpenMP/MergeSortFolder

# Compile the non-OpenMP Merge sort program
!gcc -o merge_sort Merge_Sort.c

# Compile the OpenMP Merge sort program
!gcc -fopenmp -o merge_sort_openmp OpenMP_Merge_Sort.c

# Run the non-OpenMP program and time it
import time
start_time = time.time()
!./merge_sort
end_time = time.time()
time_elapsed1 = end_time - start_time
print("Time elapsed for non-OpenMP Merge sort: ", time_elapsed1)

# Run the OpenMP program and time it
start_time = time.time()
!./merge_sort_openmp
end_time = time.time()
time_elapsed2 = end_time - start_time
print("Time elapsed for OpenMP Merge sort: ", time_elapsed2)

# Define the data for the graph
labels = ['Non-OpenMP', 'OpenMP']
times = [time_elapsed1, time_elapsed2]

# Create a bar plot of the data
x_pos = np.arange(len(labels))
plt.bar(x_pos, times)
plt.xticks(x_pos, labels)
plt.ylabel('Time (seconds)')
plt.title('Merge Sort Comparison')

# Display the plot
plt.show()
```
